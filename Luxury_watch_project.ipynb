```python
import numpy as np
watch.rename(columns={'Power Reserve':'Power Reserve (days)'},inplace=True)
watch['Power Reserve (days)']
watch['Power Reserve (days)']=watch['Power Reserve (days)'].replace('4,200',pd.NA)
```

```python
def conv_to_day(value):
   if pd.notnull(value):
    value_str=str(value)
    if 'hours' in value_str:
      return int(value.split()[0])/24
    elif 'days' in value_str:
      return int(value.split()[0])
   return value
watch['Power Reserve (days)']=watch['Power Reserve (days)'].apply(conv_to_day)


mean_power=watch['Power Reserve (days)'].mean()
watch['Power Reserve (days)'].fillna(mean_power,inplace=True)
watch['Power Reserve (days)']=watch['Power Reserve (days)'].round(2)
watch['Power Reserve (days)']
```

```python
from sklearn .model_selection import cross_val_score
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
#x=watch.drop(labels=['Power Reserve (days)','Price (USD)'],axis=1)
x=watch.drop('Price (USD)',axis=1)

y=watch['Price (USD)']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

numeric_features=x.select_dtypes(include=['float64','int64']).columns
cat_features=x.select_dtypes(include=['object']).columns

numeric_transformer=StandardScaler()
cat_transformer=OneHotEncoder(handle_unknown='ignore')

preprocessor=ColumnTransformer(transformers=[
    ('num',numeric_transformer,numeric_features),
    ('cat',cat_transformer,cat_features)

])

from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor

lr=LinearRegression()
kn=KNeighborsRegressor()
tree=DecisionTreeRegressor()

models=[lr,kn,tree]
for model in models:
  pipeline=Pipeline(steps=[('preprocessor',preprocessor),('model',model)])
  pipeline.fit(x_train,y_train)
  y_pred=pipeline.predict(x_test)
  r2=r2_score(y_test,y_pred)
  mse=mean_squared_error(y_test,y_pred)
  print('Model:{},R_Squared(R^2):{}'.format(model,r2))



```
